```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Settings

```python
sns.set_style("white")
colors = ['#1b9e77','#d95f02','#7570b3','#e7298a','#66a61e','#e6ab02']
#flatui = ['#7fc97f','#beaed4','#fdc086','#ffff99','#386cb0','#f0027f']
sns.set_palette(colors)
```

```python
filter_dates = False
```

```python
x_size, y_size = 12,8
plt.rcParams.update({'font.size': 16})
```

## Data import

```python
df = pd.read_csv("data\\counters.csv")
```

## Filter dates

```python
if filter_dates:
    date1 = "2020-10-05"
    date2 = "2020-11-08"

    dates_to_remove = ["2020-10-19", "2020-10-23", "2020-10-24", "2020-10-31"]
else:
    date1 = ""
    date2 = ""
    dates_to_remove = []
```

```python
if date1:
    df = df[df['date']>=date1]
    
if date2:
    df = df[df['date']<=date2]
    
for dtr in dates_to_remove:
    #df = df[df['date'].dt.date.astype(str) != dtr]
    df = df[df['date'] != dtr]
```

```python
counter_ids = df.counter_id.unique()
```

## Violinplots

If correct means should equal approximately `0.7142857142857143`

```python
df['workday'].describe()
```

## All

```python
for i in range(0, len(counter_ids), 5):
    df_small = df[df.counter_id.map(lambda x: x in counter_ids[i:i+5])]
    sns.catplot(data=df_small, y='count', x='counter_id', kind="violin")
    f = plt.gcf()
    f.set_size_inches(x_size,y_size)
    plt.xlabel("Counter id")
    plt.ylabel("Count [$h^{-1}$]")
    plt.xticks
    if filter_dates:
        plt.savefig("figs\\counts\\distribution_counts_"+str(i)+".pdf", bbox_inches="tight")
        plt.savefig("figs\\counts\\distribution_counts_"+str(i)+".svg", bbox_inches="tight")
        plt.savefig("figs\\counts\\distribution_counts_"+str(i)+".png")
    else:
        plt.savefig("figs\\counts_no_filter\\distribution_counts_"+str(i)+".pdf", bbox_inches="tight")
        plt.savefig("figs\\counts_no_filter\\distribution_counts_"+str(i)+".svg", bbox_inches="tight")
        plt.savefig("figs\\counts_no_filter\\distribution_counts_"+str(i)+".png")
    
```

## Workdays

```python
for i in range(0, len(counter_ids), 5):
    df_small = df[df.counter_id.map(lambda x: x in counter_ids[i:i+5])]
    sns.catplot(data=df_small, y='count', x='counter_id', hue="workday", kind="violin")
    f = plt.gcf()
    f.set_size_inches(x_size,y_size)
    plt.xlabel("Counter id")
    plt.ylabel("Count [$h^{-1}$]")
    plt.xticks
    if filter_dates:
        plt.savefig("figs\\counts\\distribution_counts_workdays_"+str(i)+".pdf", bbox_inches="tight")
        plt.savefig("figs\\counts\\distribution_counts_workdays_"+str(i)+".svg", bbox_inches="tight")
        plt.savefig("figs\\counts\\distribution_counts_workdays_"+str(i)+".png")
    else:
        plt.savefig("figs\\counts_no_filter\\distribution_counts_workdays_"+str(i)+".pdf", bbox_inches="tight")
        plt.savefig("figs\\counts_no_filter\\distribution_counts_workdays_"+str(i)+".svg", bbox_inches="tight")
        plt.savefig("figs\\counts_no_filter\\distribution_counts_workdays_"+str(i)+".png")
    
```
